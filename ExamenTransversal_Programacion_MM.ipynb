```python
from datetime import date
from tabulate import tabulate

#DECLARO EL TOTAL DE PISOS, EL TOTAL DE DEPARTAMENTOS Y SUS PRECIOS, EL DESPLIEGUE SE REALIZARÁ MOSTRANDO TODOS LOS DEPARTAMENTOS POR TIPO Y PISP
total_pisos = 10
deptos_por_piso = 4
precios = {'A': 3800, 'B': 3000, 'C': 2800, 'D': 3500}
vendidos = [['' for _ in range(deptos_por_piso)] for _ in range(total_pisos)]
compradores = {}

#CREO UNA FUNCION PARA MOSTRAR EL MENU, DESPLIEGO LAS OPCIONES DEL MENÚ
def mostrar_menu():
    print(" Bienvenido a Casa Feliz :)")
    print("******* MENÚ *******")
    print("1. Comprar departamento")
    print("2. Mostrar departamentos disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")

#DECLARO FUNCIÓN PARA COMPRAR UN DEPARTAMENTO, SOLICITO INGRESAR NUMERO DE DEPARTAMENTO Y TIPO DE DEPARTAMENTO, EN CASO DE QUE EL DPTO ESTÉ DISPONIBLE, SOLICITO EL RUT DEL COMPRADOR, EN CASO DE QUE NO, CANCELO LA OPERACION
def comprar_departamento():
    piso = int(input("Ingrese el Número de Piso (1-10): "))
    tipo = input("Ingrese el Tipo de Departamento (A, B, C o D): ")
#REALIZO VALIDACIONES AL TIPO DE DEPARTAMENTO Y AL PISO, SI NO EXISTE ALGUNO DE LOS DOS, ENVIARÁ UN MENSAJE INDICANDO QUE UNO DE LOS DOS CAMPOS INGRESADOS NO EXISTE
    if piso < 1 or piso > total_pisos or tipo.upper() not in precios:
        print("El Tipo de Departamento o el Piso Ingresado Son Invalidos")
        return

    fila = piso - 1
    columna = ord(tipo.upper()) - ord('A')
    # VALIDO QUE EL DEPARTAMENTO SELECCIONADO SE ENCUENTRE DISPONIBLE, SI SE ENCUENTRA VENDIDO DESPLEGARÁ UN MENSAJE DE ADVERTENCIA
    if vendidos[fila][columna] != '':
        print("El Departamento Seleccionado NO Está Disponible")
        return
    #SI EL DEPTO SELECCIONADO ESTÁ DISPONIBLE, SOLICITO EL RUT DEL CLIENTE, VALIDANDO QUE NO INGRESE DATOS NO SOLICITADOS, ADEMÁS, REALIZO EL MARCAJE DEL DEPTO COMPRADO
    rut = input("Ingrese el Rut del comprador (no ingrese guiones ni puntos): ")
    compradores[rut] = (tipo.upper(), piso)
    vendidos[fila][columna] = 'X'
    print("Felicitaciones! El Departamento Seleccionado Ha Sido Comprado!")

#DECLARO FUNCIÓN PARA EL DESPLIEGUE DE LOS DEPARTAMENTOS, TRANSFORMO LA VARIABLE INDICADA EN EL PRINCIPIO DONDE DECLARO LA CANTIDAD DE PISOS Y DEPTOS MOSTRANDOLA COMO TABLA
def mostrar_departamentos_disponibles():
    filas = []
    encabezado = ["Piso"] + [chr(ord('A') + j) for j in range(deptos_por_piso)]

    for i in range(total_pisos):
        fila = [str(i + 1)] + ['' if vendidos[i][j] == '' else 'X' for j in range(deptos_por_piso)]
        filas.append(fila)

    print(tabulate(filas, headers=encabezado, tablefmt="grid"))


#DECLARO FUNCION PARA VER EL LISTADO DE LOS COMPRADORES REGISTRADOS EN EL SISTEMA, ADEMÁS CONCATENO LA INFORMACIÓN DEL PISO Y EL TIPO COMPRADO, MOSTRANDOLO COMO TABLA
def ver_listado_compradores():
    print("Listado de Propietarios:")
    print("-" * 30)
    print("RUT\t Tipo\t Piso")
    print("-" * 30)

    for rut in sorted(compradores):
        tipo, piso = compradores[rut]
        print(f"{rut}\t{tipo}\t{piso}")

    print("-" * 30)

#DECLARO FUNCION DONDE REALIZO EL DESPLIEGUE DEL TOTAL DE LAS VENTAS
def mostrar_ganancias_totales():
    total = sum(precios[tipo] for tipo, _ in compradores.values())
    print("")
    print(f"Las Ventas Totales Han Sido De: {total} UF")

#DECLARO FUNCION DONDE REALIZO LA SALIDA DEL SISTEMA, DESPLEGANDO UN MENSAJE CON MI NOMBRE Y LA FECHA ACTUAL
def salida():
    fecha_actual = date.today().strftime("%d/%m/%Y")
    print(f"Adios! Se despide Matías Marín el día: {fecha_actual}")

# EJECUTO EL MENÚ DEL SISTEMA, ESTA ES LA FUNCIÓN PRINCIPAL Y REALIZA EL LLAMADO A LAS DEMÁS FUNCIONES
while True:
    mostrar_menu()
    opcion = input("Ingrese una opción Del Menú: ")

    if opcion == '1':
        comprar_departamento()
    elif opcion == '2':
        mostrar_departamentos_disponibles()
    elif opcion == '3':
        ver_listado_compradores()
    elif opcion == '4':
        mostrar_ganancias_totales()
    elif opcion == '5':
        salida()
        break
    else:
        print("La Opción Ingresada es Inválida")
```
